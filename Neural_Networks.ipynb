```python
import numpy as np
import matplotlib.pyplot as plt

line = np.linspace(-3,3, 100)
plt.plot(line, np.tanh(line), label="tanh")
plt.plot(line, np.maximum(line, 0), label="relu")
plt.legend(loc="best")
plt.xlabel("x")
plt.ylabel("tanh versus relu")
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from sklearn.inspection import DecisionBoundaryDisplay
X, y = make_moons(n_samples=100, noise=0.3, random_state=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y,
                                                    random_state=2)
```

```python
mlp = MLPClassifier(solver="lbfgs", alpha=0.0001).fit(X_train, y_train)
```

```python
X_train[:,0]
```

```python
DecisionBoundaryDisplay.from_estimator(mlp, X_train)
plt.scatter(X_train[:,0], X_train[:,1], c=y_train, cmap=plt.cm.RdYlBu)
```

```python
y_train
```

```python
# using two hidden layders with 10 units each
mpl = MLPClassifier(solver="lbfgs", random_state=1, hidden_layer_sizes=[10,10])
mpl.fit(X_train, y_train)
```

```python
DecisionBoundaryDisplay.from_estimator(mlp, X_train)
plt.scatter(X_train[:,0], X_train[:,1], c=y_train, cmap=plt.cm.RdYlBu)
```

```python
fig, axes = plt.subplots(2, 4, figsize=(20, 8))
for i, ax in enumerate(axes.ravel()):
  mpl = MLPClassifier(solver="lbfgs", random_state=i, hidden_layer_sizes=[100, 100])
  mpl.fit(X_train, y_train)
  DecisionBoundaryDisplay.from_estimator(mlp, X_train, ax=ax)
  ax.scatter(X_train[:,0], X_train[:,1], c=y_train, cmap=plt.cm.RdYlBu)
plt.show()
```

```python
fig, axes = plt.subplots(2, 4, figsize=(20,8))
for axx, hidden_layer in zip(axes, [10, 100]):
  print(ax)
  print(hidden_layer)
  for ax, alpha in zip(axx, [0.0001, 0.01, 0.1, 1]):
    mpl = MLPClassifier(solver="lbfgs", random_state=0, hidden_layer_sizes=[hidden_layer, hidden_layer],
                        alpha=alpha)
    mpl.fit(X_train, y_train)
    DecisionBoundaryDisplay.from_estimator(mlp, X_train, ax=ax)
    ax.scatter(X_train[:,0], X_train[:,1], c=y_train, cmap=plt.cm.RdYlBu)
    ax.set_title(f"{hidden_layer, hidden_layer, alpha}")
```

```python
from sklearn.datasets import load_breast_cancer
X, y = load_breast_cancer().data, load_breast_cancer().target
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=0)
```

```python
mlp = MLPClassifier(random_state=2)
mlp.fit(X_train, y_train)
print(f"{mlp.score(X_train, y_train)}")
print(f"{mlp.score(X_test, y_test)}")
```

```python
train_mean = X_train.mean(axis=0)
train_std = X_train.std(axis=0)
X_train_scaled = (X_train - train_mean)/train_std
X_test_scaled = (X_test - train_mean)/train_std
```

```python
mlp = MLPClassifier(random_state = 40)
mlp.fit(X_train_scaled, y_train)
```

```python
print(mlp.score(X_train_scaled, y_train))
print(mlp.score(X_test_scaled, y_test))
```
